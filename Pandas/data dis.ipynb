```python
from pandas import Series, DataFrame
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/Users/pengfei/Downloads/diary.txt', header = None, names=["count_c", "diary_id"])
```

```python
data.count_c.describe()
```

```python
data['ctr_value']
```

```python
tt = data["ctr_value"].hist(bins=20, alpha=0.5)
plt.show(tt)
```

```python
quartiles = pd.cut(data["ctr_value"], 100)
grouped = data.ctr_value.groupby(quartiles)

def get_stats(group):
    return {'ctr_values': group.count()}

grouped.apply(get_stats).unstack()
```

```python
quartiles = pd.cut(data["org_deal_value"], 20)
grouped = data.ctr_value.groupby(quartiles)

def get_stats(group):
    return {'org_deal_values': group.count()}

grouped.apply(get_stats).unstack()
```

```python
quartiles = pd.cut(data["org_income_value"], 20)
grouped = data.ctr_value.groupby(quartiles)

def get_stats(group):
    return {'income_values': group.count()}

grouped.apply(get_stats).unstack()
```
