```python
import sys
import re

import pandas as pd
import numpy as np
from pandas import Series, DataFrame, Panel

# parse nginx log [xx] [xx]
pat = re.compile(r'\[(.*?)\]')
data = []
for i in open('data/profiler.log'):
    _dl = pat.findall(i)
    if _dl: data.append(_dl)

total = []
for d in data:
    if len(d) < 23: continue
    b = {}
    b['access time'] = d[0]
    b['status'] = d[1]
    b['process time'] = d[2]
    b['method'] = d[3]
    b['uri'] = d[4]
    b['branding'] = d[5]
    b['version'] = d[6]
    b['phone'] = d[7]
    b['manufacturer'] = d[8]
    b['first_install_time'] = d[-2]
    b['first_launch_time'] = d[-1]
    total.append(b)

df = DataFrame(total)
df['status'] = df['status'].astype('int')
df['status'].plot()
```

```python
df.index = pd.to_datetime(df.pop('access time'))
```

```python
dfgrouped = df.groupby(df['status'])
dfgrouped.size().plot(kind='bar')
```

```python
dfgrouped = df.groupby(df['method'])
dfgrouped.size().plot(kind='bar')
```

```python
dfgrouped = df.groupby(['branding'])
dfgrouped.size().plot(kind='bar')
```
